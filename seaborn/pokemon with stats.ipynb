```python
pkmn_stats = pd.melt(frame=pkmn, id_vars=['Name', 'Type 1', 'Type 2'], var_name='Stat')
```

```python
pkmn_stats.head()
```

```python
pkmn_stats.info()
```

```python
plt.figure(figsize=(15, 10))
plt.ylim(0, 275)
sns.swarmplot(x='Stat', y='value', data=pkmn_stats, hue='Type 1', dodge=True, size=7)
plt.legend(loc=2, bbox_to_anchor=(1,1), borderaxespad=0.)
plt.show()
```

```python
sns.set_style('whitegrid')
with sns.color_palette([
    "#8ED752", "#F95643", "#53AFFE", "#C3D221", "#BBBDAF",
    "#AD5CA2", "#F8E64E", "#F0CA42", "#F9AEFE", "#A35449",
    "#FB61B4", "#CDBD72", "#7673DA", "#66EBFF", "#8B76FF",
    "#8E6856", "#C3C1D7", "#75A4F9"], n_colors=18, desat=.9):
    plt.figure(figsize=(15, 10))
    plt.ylim(0, 275)
    sns.swarmplot(x='Stat', y='value', data=pkmn_stats, hue='Type 1', dodge=True, size=7)
    plt.legend(loc=2, bbox_to_anchor=(1,1), borderaxespad=0.)
    plt.show()
```
